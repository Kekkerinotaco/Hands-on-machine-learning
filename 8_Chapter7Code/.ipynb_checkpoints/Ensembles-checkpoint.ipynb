# Building an ensemble classifier

```python
from sklearn.datasets import make_moons
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
moons = make_moons(n_samples=10000, noise=0.4)
```

```python
from sklearn.model_selection import train_test_split

x = moons[0]
y = moons[1]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
log_clf = LogisticRegression()
rnd_clf = RandomForestClassifier()
svm_clf = SVC(probability=True)

voting_clf = VotingClassifier(
    estimators=[("lr", log_clf), ("rf", rnd_clf), ('svc', svm_clf)], 
    voting="soft"
)

voting_clf.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

for clf in (log_clf, rnd_clf, svm_clf, voting_clf):
    clf.fit(x_train, y_train)
    y_hat = clf.predict(x_test)
    print(clf, accuracy_score(y_test, y_hat))
```

# Bagging and Pasting

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

bag_clf = BaggingClassifier(DecisionTreeClassifier(),
                            n_estimators=500, 
                            max_samples=100, 
                            bootstrap=True,
                            n_jobs=-1, 
                            oob_score=True)

bag_clf.fit(x_train, y_train)
y_pred = bag_clf.predict(x_test)
```

```python
tree_clf = DecisionTreeClassifier(max_depth=30)
tree_clf.fit(x_train, y_train)
y_pred_tree = tree_clf.predict(x_test)
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

def plot_results(x, y, title):
    fig, axs = plt.subplots()
    plt.scatter(x[:, 0], x[:, 1], c=y)
    axs.set_xlabel("X1")
    axs.set_ylabel("X2")
    axs.set_title(title)
```

```python
plot_results(x_test, y_test, title="True distribution")
plot_results(x_test, y_pred, title="Bagging Clf")
plot_results(x_test, y_pred_tree, title="Tree Clf")
```

# Estimating relative features importance with RandomForest

```python
from sklearn.datasets import load_iris

iris=load_iris()
rnd_clf = RandomForestClassifier(n_estimators=500, n_jobs=-1)
rnd_clf.fit(iris['data'], iris['target'])
for name, score in zip(iris["feature_names"], rnd_clf.feature_importances_):
    print(name, score)
```

# Implementing early stopping with DecisionTrees

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
gbrt = GradientBoostingRegressor(max_depth=2, 
                                 n_estimators=120)

gbrt.fit(x_train, y_train)
```

```python
errors = [mean_squared_error(y_test, y_hat) for y_hat in gbrt.staged_predict(x_test)]
```

```python
best_n_estimators = np.argmin(errors) + 1
best_n_estimators
```

```python
gbrt_best = GradientBoostingRegressor(max_depth=2, n_estimators=best_n_estimators)
gbrt_best.fit(x_train, y_train)
```

```python
gbrt_best.fit(x_train, y_train)
y_hat = gbrt_best.predict(x_train)
mean_squared_error(y_train, y_hat)
```

```python
gbrt_best.fit(x_train, y_train)
y_hat = gbrt_best.predict(x_test)
mean_squared_error(y_test, y_hat)
```

## Early stopping with warm start

```python
gbrt = GradientBoostingRegressor(max_depth=2, warm_start=True)

min_pred_error = float("inf")
error_going_up = 0
for n_estimators in range(1, 121):
    gbrt.n_estimators = n_estimators
    gbrt.fit(x_train, y_train)
    y_hat = gbrt.predict(x_test)
    pred_error = mean_squared_error(y_test, y_hat)
    if pred_error < min_pred_error:
        min_pred_error = pred_error
        error_going_up = 0
    else:
        error_going_up +=1
        if error_going_up == 5:
            gbrt.n_estimators = gbrt.n_estimators - 5
            break
```
