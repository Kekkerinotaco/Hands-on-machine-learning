# Working with blobs data

```python
from sklearn.datasets import make_blobs
import numpy as np

blob_centers = np.array(
    [[ 0.2,  2.3],
     [-1.5 ,  2.3],
     [-2.8,  1.8],
     [-2.8,  2.8],
     [-2.8,  1.3]])

blob_std = np.array([0.4, 0.3, 0.1, 0.1, 0.1])
```

```python
x, y = make_blobs(n_samples=2000, centers=blob_centers, 
                  cluster_std=blob_std, random_state=42)
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

def plot_clusters(x):
    plt.scatter(x[:, 0], x[:, 1])
    plt.xlabel("X")
    plt.ylabel("Y")
    
plot_clusters(x)
```

```python
from sklearn.cluster import KMeans

k = 5
kmean = KMeans(n_clusters=k)
y_hat = kmean.fit_predict(x)
```

```python
kmean.labels_
```

```python
kmean.cluster_centers_
```

```python
kmean.inertia_
```

```python
def plot_inertia(x, n_centers=[]):
    inertias = []
    for center in n_centers:
        kmeans = KMeans(n_clusters=center)
        kmeans.fit(x)
        inertia = kmeans.inertia_
        inertias.append(inertia)
        
    fig, ax = plt.subplots(figsize=(12, 8))
    ax.plot(n_centers, inertias, marker="o")
    ax.set_xlabel("Number of clusters")
    ax.set_ylabel("Inertia")

centers = [x for x in range(1, 20)]
plot_inertia(x, centers)
```

```python
kmean.labels_
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(x, kmean.labels_)
```

# Image segmentation

```python
from matplotlib.image import imread
```

```python
def draw_image(image, colors=[]):
    n_color_index = 0
    fig, axs = plt.subplots(2, 4, figsize=(16, 12))
    fig.subplots_adjust(wspace=0.05, hspace=0.01)
    x = image.reshape(-1, image.shape[2])
    for row in range(2):
        for column in range(4):
            kmeans = KMeans(n_clusters=colors[n_color_index]).fit(x)
            segmented_image = kmeans.cluster_centers_[kmeans.labels_]
            segmented_image = segmented_image.reshape(image.shape)
            axs[row, column].imshow(segmented_image)
            axs[row, column].set_title("Number of colors: {}".format(colors[n_color_index]))
            axs[row, column].axis("off")
            n_color_index+= 1
    plt.savefig("Revoked.jpeg")
```

```python
image_path = r"Снимок.PNG"
image = imread(image_path)
n_colors = [2, 4, 6, 8, 10, 12, 14, 16]
draw_image(image, n_colors)
```

# Using Clustering for Preprocessing

```python
from sklearn.datasets import load_digits

x_digits, y_digits = load_digits(return_X_y = True)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x_digits, y_digits, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(multi_class="ovr", solver = "lbfgs", 
                             max_iter=5000, random_state=42)

log_reg.fit(X_train, y_train)
```

```python
log_reg_score = log_reg.score(X_test, y_test)
log_reg_score
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

pipeline = Pipeline([
    ("kmeans", KMeans(n_clusters=50)),
    ("STDscaler", StandardScaler()),
    ("log_reg", LogisticRegression())
])

pipeline.fit(X_train, y_train)
```

```python
pipeline.score(X_test, y_test)
```

### Finding the best number of clusters via GridSearch. Combining GridSearch and PipeLine

```python
from sklearn.model_selection import GridSearchCV

param_grid = dict(kmeans__n_clusters=np.arange(10, 100, 10), 
                  log_reg__penalty=["l1", "l2"], 
                  log_reg__C=np.arange(0.1, 1, 0.2))
grid_clf = GridSearchCV(pipeline, param_grid, cv=3, verbose=1)
grid_clf.fit(X_train, y_train)
```

```python
grid_clf.best_params_
```

```python
grid_clf.score(X_test, y_test)
```

```python
# Trying another way of implementing param_grid

param_grid = {"kmeans__n_clusters":np.arange(10, 100, 10), 
             "log_reg__penalty":["l1", "l2"], 
             "log_reg__C":np.arange(0.1, 1, 0.2)}
grid_clf = GridSearchCV(pipeline, param_grid, cv=3, verbose=1)
grid_clf.fit(X_train, y_train)
```

```python
grid_clf.score(X_test, y_test)
```

# Training a DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=1000, noise=0.05)
dbscan = DBSCAN(eps=0.05, min_samples=5)
dbscan.fit(X)
```

```python
def plot_moons(x, y, y_hat):
    fig, axs = plt.subplots(1, 2, figsize=(18, 9), layout="constrained")
    axs[0].scatter(x[:, 0], x[:, 1], c=y)
    axs[1].scatter(x[:, 0], x[:, 1], c=y_hat)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y, dbscan.labels_)
```

```python
plot_moons(X, y, dbscan.labels_)
```

```python
X, y = make_moons(n_samples=1000, noise=0.05)
dbscan = DBSCAN(eps=0.2, min_samples=5)
dbscan.fit(X)
```

```python
accuracy_score(y, dbscan.labels_)
```

```python
plot_moons(X, y, dbscan.labels_)
```

```python
dbscan.components_
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=50)
knn.fit(dbscan.components_, dbscan.labels_[dbscan.core_sample_indices_])
```

```python
X_new = np.array([[-0.05, 0], 
                   [0, 0.5]])

knn.predict(X_new)
```

```python
knn.predict_proba(X_new)
```
